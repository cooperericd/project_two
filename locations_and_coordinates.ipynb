```python
#Dependencies and Setup
import pandas as pd

#File to Load
world_cities_csv_file = "worldcities.csv"
us_states_csv_file = "50_us_states_all_data.csv"

#Read cities file and store in Pandas data frame
city_data = pd.read_csv(world_cities_csv_file)
state_data = pd.read_csv(us_states_csv_file)
```

```python
city_data.head()
```

```python
state_data.head()
```

```python
#merge city and state dataframes
merge_df = pd.merge(city_data, state_data, how="left", left_on="admin_name", right_on="State")
```

```python
merge_df.head()
```

```python
#filter by city
new_york = merge_df.loc[merge_df['city']=='New York']
new_york
```

```python
#filter by city and iso2
bogota = merge_df.loc[merge_df['city']=='Bogota'].loc[merge_df['iso2']=='CO']
bogota                
```
